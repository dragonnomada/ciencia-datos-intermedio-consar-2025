```python
import pandas

pandas.__version__
```

```python
edades = pandas.Series([
    24, 25, 65, 54, 45, 
    34, 43, 21, 13, 18, 
    92, 36, 38, 21, 56,
    32, 24, 28, 34, 37
])

edades
```

```python
edades.describe()
```

```python
edades.mean()
```

```python
print("Intervalo de confianza al 68%")
print(f"{edades.mean() - edades.std():.2f}, {edades.mean() + edades.std():.2f}")
```

```python
print("Intervalo de confianza al 95%")
print(f"{edades.mean() - 2 * edades.std():.2f}, {edades.mean() + 2 * edades.std():.2f}")
```

```python
print("Intervalo de confianza al 99.7%")
print(f"{edades.mean() - 3 * edades.std():.2f}, {edades.mean() + 3 * edades.std():.2f}")
```

```python
edades.plot.hist()
```

## Caso de estudio 1 - Estatura contra Peso

```python
estaturas = pandas.Series([
    1.78, 1.63, 1.54, 1.81, 1.67, 
    1.62, 1.89, 1.71, 1.70, 1.68,
    1.56, 1.53, 1.48, 1.52, 1.71,
    1.85, 1.72, 1.63, 1.56, 1.78,
])

estaturas.describe()
```

```python
pesos = pandas.Series([
    88, 72, 54, 90, 72,
    56, 100, 90, 94, 80,
    57, 64, 50, 65, 72,
    74, 78, 58, 48, 76
])

pesos.describe()
```

```python
estaturas.plot.density()
```
